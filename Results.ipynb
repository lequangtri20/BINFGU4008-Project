```python
def repeat_elements(input_list, n):
    return [item for item in input_list for _ in range(n)]
```

```python
def extract_mean_and_ci(row):

  return f'{row[1]:.2f} ({row[0]:.2f}, {row[2]:.2f})'
```

```python
resnet18_total = pd.concat((pd.DataFrame(resnet18_imagenet_no_gender).T, pd.DataFrame(resnet18_imagenet_w_gender).T, pd.DataFrame(resnet18_xavier_no_gender).T,
                            pd.DataFrame(resnet18_xavier_no_gender_10).T, pd.DataFrame(resnet18_xavier_w_gender).T, pd.DataFrame(resnet18_xavier_w_gender_10).T), axis = 0)
resnet50_total = pd.concat((pd.DataFrame(resnet50_imagenet_no_gender).T, pd.DataFrame(resnet50_imagenet_w_gender).T, pd.DataFrame(resnet50_xavier_no_gender).T,
                            pd.DataFrame(resnet50_xavier_no_gender_10).T, pd.DataFrame(resnet50_xavier_w_gender).T, pd.DataFrame(resnet50_xavier_w_gender_10).T), axis = 0)
alexnet_total = pd.concat((pd.DataFrame(alexnet_imagenet_no_gender).T, pd.DataFrame(alexnet_imagenet_w_gender).T, pd.DataFrame(alexnet_xavier_no_gender).T,
                           pd.DataFrame(alexnet_xavier_no_gender_10).T, pd.DataFrame(alexnet_xavier_w_gender).T, pd.DataFrame(alexnet_xavier_w_gender_10).T), axis = 0)
vgg16_total = pd.concat((pd.DataFrame(vgg16_imagenet_no_gender).T, pd.DataFrame(vgg16_imagenet_w_gender).T, pd.DataFrame(vgg16_xavier_no_gender).T,
                         pd.DataFrame(vgg16_xavier_no_gender_10).T, pd.DataFrame(vgg16_xavier_w_gender).T, pd.DataFrame(vgg16_xavier_w_gender_10).T), axis = 0)

all_results = pd.concat((resnet18_total, resnet50_total, alexnet_total, vgg16_total), axis = 0)
all_results.columns = ['lower_ci', 'mean', 'upper_ci']

all_results = all_results.reset_index(drop = False)

all_results['model'] = repeat_elements(['ResNet-18', 'ResNet-50', 'AlexNet', 'VGG-16-BN'], 30)
all_results['gender'] = repeat_elements(['No', 'Yes', 'No', 'No', 'Yes', 'Yes'], 5) * 4
all_results['weights'] = repeat_elements(['ImageNet', 'ImageNet', 'Xavier', 'Xavier', 'Xavier', 'Xavier'], 5) * 4
all_results['epochs'] = repeat_elements([7, 7, 7, 10, 7, 10], 5) * 4

all_results['index'] = all_results['index'].replace({
    'r2': 'R2',
    'pearson': 'Pearson',
    'mse': 'MSE',
    'mae': 'MAE',
    'mape': 'MAPE'
})

all_results.columns = ['Metric', 'Lower_CI', 'Mean', 'Upper_CI', 'Model', 'Gender', 'Init_Weights', 'Num_Epochs']
all_results = all_results[['Model', 'Init_Weights', 'Gender', 'Num_Epochs', 'Metric', 'Lower_CI', 'Mean', 'Upper_CI']]

all_results.to_csv('/content/drive/MyDrive/BINF_4008_Final_Project/Tables/performance_summary_table.csv', index = False)
```
